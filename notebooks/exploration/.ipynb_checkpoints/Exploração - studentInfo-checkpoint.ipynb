## Estudantes individuais

Vemos que o número de estudantes unicos é menor que a quantidade de linhas da tabela, o que significa que o mesmo estudante pode ter se matriculado em mais de um curso ou mais de uma oferta.

```python
unique_students = data['studentInfo'].id_student.unique().shape[0]
table_len = len(data['studentInfo'])

print(f"Total de estudantes: {unique_students}\nLinhas na tabela: {table_len}")
```

```python
counts = data['studentInfo'].id_student.value_counts()
student_more_then_once = np.array(counts[counts > 1].index)

student_more_then_once
```

## Mudanças durante a repetição de matricula

Dentre os alunos que se matricularam em mais de um curso ao longo do intervalo em que os dados foram coletados, observamos que houveram poucas alterações nas informações pessoais dos alunos entre as matriculas (genero, região, educação, IMD, deficiencias).

As mudanças entre matriculas se concentraram em variáveis em que seria obvias:
  - **code_presentation**: Como cada curso é apresentado em um semestre/ano diferente, era de se esperar que essa variável fosse a que sofresse alterações na maioria dos estudantes que se matricularam mais de uma vez.
  - **code_module**: Essa variavel vai sofrer alterações sempre que o aluno se rematricular em cursos diferentes. Os casos em que essa variável não mudou representam situações em que o individuo repetiu o mesmo curso.
  - **final_result**: Se o estudante teve variações nesse campo pode ter sido reprovado e depois conseguiu passar, ou então obteve distinção em um dos cursos e no outro passou com nota regular, outro caso possivel seria o aluno que desistiu de fazer o curso e depois se matriculou novamente e foi até o final. 
  - **num_of_prev_attempts**: Esse campo se altera quando o individuo faz mais de uma vez o mesmo módulo. Portanto os estudantes que não possuem duplicação nesse campo tentaram módulos diferentes, não repetindo o mesmo modulo.
  - **studied_credits**: A quantidade de créditos em que o estudante está matriculado se altera caso ele esteja fazendo uma quantidade diferente de créditos no ato da matricula.
  - **age_band**: Com bem menos individuos com mudança nesse atributo, significa que a pessoa fez aniversário entre as matriculas e possuia 35 ou 55 anos o que fez com que sua faixa de idade se alterasse.
  

```python
repeated_student_info = data['studentInfo'][data['studentInfo'].id_student.isin(student_more_then_once)]

def check_changed(df):
    return pd.DataFrame({
        "column_changed": [ col for col in df.columns if len(df[col].unique()) != 1]
    })

changes = repeated_student_info.groupby("id_student").apply(check_changed)

plt.figure(figsize=(4, 3))
changes.column_changed.value_counts().plot.barh()
plt.xlabel("Number of students");
```

#### IMD 

IMD, sigla para *Index of Multiple Deprivation* é um indice utilizado no Reino Unido, que mede o grau de "pobreza" de uma região[[1].](https://en.wikipedia.org/wiki/Multiple_deprivation_index)

No gráfico a seguir vemos que dentro do dataset observado, o indice possui sua maior frequencia na faixa "20-30%", valor relativamente baixo. A pesar disso vemos uma grande frequencia em todas as faixas, até mesmo nas mais pobres ("90-100%").


```python
imd_band = data['studentInfo']\
    .imd_band\
    .map(lambda x: '10-20%' if '10-20' == x else x)\
    .value_counts()\
    .sort_index(ascending=False)

plt.figure(figsize=(4, 3))
imd_band.plot.barh()
plt.xlabel("Number of students");
```

### Faixa de idade

Vemos que a grande maioria dos estudantes tem entre 0-35 anos, e poucos tem mais que 55.

```python
age_band = data['studentInfo'].age_band.value_counts().sort_index(ascending=False)
plt.figure(figsize=(3, 2))
age_band.plot.barh()
plt.xlabel("Number of students");
```

### Resultado final

Vemos que a quantidade de pessoas nas categorias "negativas" *Fail* e *Withdrawn* totalizam uma quantidade maior do que o total nas categorias "positivas" *Pass* e *Distinction*.

```python
final_result = data['studentInfo'].final_result.value_counts().iloc[[1, 2, 0, 3]]
plt.figure(figsize=(3, 2))
final_result.plot.barh(color=['orange', 'orange', 'lightgreen', 'lightgreen'])
plt.xlabel("Number of students");
```

### Disability
