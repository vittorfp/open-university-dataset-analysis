### Tabela *studentAssessment*

```python
studentAssessment = data['studentAssessment'].set_index("id_assessment").join(data['assessments'].set_index("id_assessment"))
studentAssessment
```

```python
def extract_features(df):
    return pd.Series({
        "totalAssessmentsTaken": len(df),
        "meanTMAScore": df.query("assessment_type == 'TMA'").score.dropna().mean(),
        "meanCMAScore": df.query("assessment_type == 'CMA'").score.dropna().mean(),
        "meanExamScore": df.query("assessment_type == 'Exam'").score.dropna().mean(),
        "submittedLate": sum((df.date - df.date_submitted) < 0),
        "totalBanked": sum(df.is_banked)
    })

features = studentAssessment.groupby(["id_student", "code_module", "code_presentation"]).apply(extract_features)
features
```

```python
features.to_csv('../../data/interim/studentAssessmentFeatures.csv')
```
